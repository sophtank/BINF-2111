```python
import pandas as pd
import matplotlib.pyplot as plt
#1. draw a pie chart of the five top programming languages using the csv on github.
```

```python
df = pd.read_csv('/home/stank/binf2111/lab12/Top5_trendingCPUlang.csv')
```

```python
language_data = df["Lang"]
amount = df["Amount"]
colors = ["#D88C9A" , "#F2D0A9" , "#F1E3D3" , "#99C1B9" , "#8E7DBE"]
explode = (0.1 , 0,0,0,0)
```

```python
plt.pie(amount , labels=language_data , explode=explode, colors=colors,autopct='%1.1f%%',shadow=True, startangle=140)
plt.title("Top five most popular CPU programming languages.")
```
