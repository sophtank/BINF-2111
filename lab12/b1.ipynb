```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#3. Draw a scatterplot of all the fungal genomes by GC content, genome size, protein and gene content based largest 100 genomes for fungi.
```

```python
my_types = {
"Species":"string",
"Kingdom":"string",
"Class":"string",
"Assembly status":"string",

}
```

```python
euk = pd.read_csv('/home/stank/binf2111/lab12/eukaryotes.tsv',sep="\t",dtype=my_types,na_values=["-"])
```

```python
selected1 = euk[
    (euk["Kingdom"] == "Fungi")
    & (euk["Number of genes"]>1000).fillna(False)
]
selected1[["Species","Size (Mb)", "Number of genes","Number of proteins" , "GC%"]]
```

```python
#genome size as x axis
#protein on y?
# gc content as size
#gene content as color
sns.relplot(
    data = selected1,
    x = "Size (Mb)",
    y = "Number of proteins",
    size = "Number of genes",
    sizes = (1,100),
    hue = "GC%",
)
plt.title("Size of gene v.s. number of protein for fungus genes")
```
